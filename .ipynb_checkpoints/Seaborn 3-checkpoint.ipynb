```python
cars = sns.load_dataset("mpg").dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.boxplot(cars.mpg);
```

```python
sns.boxplot(cars.acceleration);
```

```python
sns.boxplot(cars.horsepower);
```

```python
cars.mpg.describe().T
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg);
```

```python
sns.boxplot(y=cars.origin, x=cars.mpg);
```

```python
sns.boxplot(x="origin", y="acceleration", data=cars);
```

```python
sns.boxplot(x="origin", y="acceleration", 
            hue="cylinders", data=cars);
```

```python
sns.boxplot(y="origin", x="horsepower", 
            hue="cylinders", data=cars);
```

```python
cars.model_year.describe()
```

```python
cars["newer_model"]=cars.model_year>76
```

```python
sns.boxplot(x="origin", y="horsepower", 
            hue="newer_model", data=cars);
```

```python
sns.boxplot(x="origin", y="horsepower", 
            hue="newer_model",
            hue_order=[True,False], data=cars);
```

```python
sns.boxplot(x="origin", y="horsepower", 
            color="y", data=cars);
```

```python
cars.cylinders.value_counts()
```

```python
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(x="cylinders", y="displacement", data=cars);
```

```python
sns.violinplot(x="cylinders", y="displacement", data=cars,
              hue="origin");
```

```python
sns.violinplot(x="cylinders", y="displacement", 
              data=cars[cars.origin.isin(["japan", "europe"])], 
               hue="origin")
plt.legend(loc=2);
```

```python
sns.violinplot(x="cylinders", y="displacement", 
              data=cars[cars.origin.isin(["japan", "europe"])], 
               hue="origin", split=True)
plt.legend(loc=2);
```

```python
sns.violinplot(x="cylinders", y="displacement", 
              data=cars[cars.origin.isin(["japan", "europe"])], 
               hue="origin", split=True, inner="quartiles")
plt.legend(loc=2);
```

```python
sns.violinplot(x="cylinders", y="displacement", 
              data=cars[cars.origin.isin(["japan", "europe"])], 
               hue="origin", split=True, 
              scale="count")
plt.legend(loc=2);
```

```python
sns.violinplot(x="cylinders", y="displacement", hue="origin",
              data=cars[cars.origin.isin(["japan", "europe"])], 
               split=True,
               scale="count",
               scale_hue==False,
              )
plt.legend(loc=2);
```
