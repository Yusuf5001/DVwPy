```python
days = [1,2,3,4,5,6,7]
max_t = [22, 51, 52, 58, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]

```

```python
plt.plot(days, max_t)
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend()
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(loc="upper left")
```

```python
x = np.linspace(0,5,100)
y= np.sin(4*x)*np.exp(-x)
z = np.exp(-x)*np.cos(4*x)
plt.figure()
plt.plot(x,y,"r--", label="$e^{-x} sin(4x)$", linewidth=4.0)
plt.plot(x,z, label="$e^{-x} cos(4x)$", linewidth=2.0)
plt.xlabel("x")
plt.ylabel("value")
plt.legend()
plt.title("My plot")
plt.savefig("myFigure.png")
```

```python
x = np.linspace(0,10,10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y1, "ro")
plt.title("$y_1=x$")

plt.subplot(2,2,2)
plt.plot(x,y1, "g--")
plt.title("$y_2=x^2$")

plt.subplot(2,2,3)
plt.plot(x,y3, "b^")
plt.title("$y_3=x^3$")

plt.subplot(2,2,4)
plt.plot(x,y4, "ks")
plt.title("$y_4=\sqrt{x}$")
```

```python
fig, axes=plt.subplots(nrows=1, ncols=2)
for a in axes:
    if (5%2==0):
        a.plot(x,y)
    else:
        a.plot(x,y)
```
