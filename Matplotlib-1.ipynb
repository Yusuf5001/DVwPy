```python
age = [25,27,29, 31, 33, 35, 37, 39, 41, 43 ,45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62312, 66853, 67313, 68211, 73123]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age-Salary")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(age, salary, "r")
plt.subplot(2,1,2)
plt.plot(salary, age, "k")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age")
plt.show()
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(age, salary, "r")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")
ax[0].set_title("Salary by Age")
ax[1].plot(salary, age, "k")
ax[1].set_xlabel("Salary")
ax[1].set_ylabel("Age")
ax[1].set_title("Age by Salary")
plt.show()
```

```python
fig, ax = plt.subplots(2,3)

ax[0][1].plot(age, salary, "r")
ax[0][1].set_xlabel("Age")
ax[0][1].set_ylabel("Salary")
ax[0][1].set_title("Salary by Age")

ax[1][2].plot(salary, age, "k")
ax[1][2].set_xlabel("Salary")
ax[1][2].set_ylabel("Age")
ax[1][2].set_title("Age by Salary")
plt.show()
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(age, salary, "r")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Salary")
ax[0].set_title("Salary by Age")
ax[1].plot(salary, age, "k")
ax[1].set_xlabel("Salary")
ax[1].set_ylabel("Age")
ax[1].set_title("Age by Salary")
plt.show()
```

```python
salary2 = [ 66853, 67313, 68211, 73123,38496, 42000, 46752, 49320, 53200, 56000, 62312]
plt.plot(age, salary, label="age1")
plt.plot(age*2, salary*2, label="age2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Age&Salary")
plt.legend()
plt.show()
```

```python
x = np.arange(0,11)
y = x.copy()
z= x*2
t = np.log(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
ax.plot(x,t)
ax.set_xlabel("x")
ax.set_ylabel("y or z or t")
ax.set_title("...")

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(13,7))

ax[0,0].plot(age, salary, ls="-", marker="+", markersize=30)
ax[0,0].set_title("Salary by Age")

ax[0,1].plot(salary_2, age, "b", ls="-.", marker="o", markersize=30)
ax[0,1].set_title("Salary2 by Age")

ax[1,0].plot(age, salary, "g", ls=":",  marker="x", markersize=30)
ax[1,0].set_title("Age by Salary")

ax[1,1].plot(salary_2, age, ls="--", marker="*", markersize=30)
ax[1,1].set_title("Age by Salary2")

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(13,7))

ax[0][1].plot(age, salary, ls="-", marker="+", markersize=30)
ax[0,1].set_title("Salary by Age")

ax[0,0].plot(salary_2, age, "b", ls="-.", marker="o", markersize=30)
ax[0,0].set_title("Salary2 by Age")

ax[1,0].plot(age, salary, "g", ls=":",  marker="x", markersize=30)
ax[1,0].set_title("Age by Salary")

ax[1,1].plot(salary_2, age, ls="--", marker="*", markersize=30)
ax[1,1].set_title("Age by Salary2")

plt.show()
```
