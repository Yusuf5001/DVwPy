```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore") 
warnings.warn("this will not show")

%matplotlib inline

plt.rcParams["figure.figsize"] = (10,6)

sns.set_style("whitegrid") 
pd.set_option('display.float_format', lambda x: '%.3f' % x)

pd.options.display.max_rows = 150 
pd.options.display.max_columns = 100
```

```python
x_p = np.array([1,2,6,8,11])
y_p = np.array([3,5,7,10,15])
plt.plot(x_p, y_p);
```

```python
plt.xlabel("x.")
plt.ylabel("y.")
plt.title("x vs y")
plt.plot(x_p, y_p);
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("XvsY")
ax.plot(x_p, y_p);
```

```python
fig, ax = plt.subplots(1,2)
ax[0].set_title("XvsY")
ax[0].plot(x_p, y_p)

ax[1].set_title("YvsX")
ax[1].plot(y_p, x_p)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x_p, y_p)
plt.subplot(1,2,2)
plt.plot(y_p, x_p)
```
