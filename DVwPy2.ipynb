```python
company = ["Google", "Amazon", "Microsoft", "Facebook"]
revenue = [90, 136, 89, 27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company, revenue)
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(company, revenue)
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(company, revenue, label="Revenue")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(company, revenue, label="Revenue")
plt.bar(xpos, profit, label="profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos+0.2, revenue, label="Revenue")
plt.bar(xpos-0.2, profit, label="profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos+0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos-0.2, profit, width=0.4 ,label="profit")
plt.legend()
plt.show()
```

```python
plt.yticks(xpos, company)
plt.title("US Tech Stocks")

plt.barh(xpos+0.2, revenue, label="Revenue")
plt.barh(xpos-0.2, profit ,label="profit")
plt.legend()
plt.show()
```

```python
exp_values = [1400, 600, 300, 410, 250]
exp_labels = ["Rent", "Food", "Phone/Internet", "Car", "Other"]
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels, radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels, radius=2, autopct="%0.1f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels, radius=2, autopct="%0.2f%%", shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels, radius=2, autopct="%0.2f%%", shadow=True, explode=[0,0.5,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values, labels=exp_labels, radius=2, autopct="%0.2f%%", shadow=True, explode=[0,0,0.3,0,0])
plt.show() # explode goes in anticlocwisde coz labels are rent, food, phone in that order
```
