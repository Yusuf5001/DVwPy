```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,58,47,49,46]
```

```python
plt.plot(x,y, color="red", linewidth=4)
```

```python
plt.plot(x,y, color="red", linewidth=4, linestyle="dashdot")
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color="purple", linewidth=4, linestyle="dashdot")
```
