```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age,salary)
plt.show()
```

```python
plt.plot(salary, age, color="green", marker="+")
plt.xlabel("Salary")
plt.ylabel("Age")
plt.title("Salary-Age Comparison")
plt.show()
```

```python
# Functional Method plt.plot(x,y)
# Object Oriented method
```

```python
fig, ax = plt.subplots() # fig is container  and ax = list of individual charts
ax .plot(age,salary, "r")  # r is red colour and default is blue.
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Salary by age")
```

```python
x = np.arange(0,10)
x
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x,x+2, color="black", lw=.25, linestyle="-")
ax.plot(x,x*2, color="blue", lw=.5, linestyle="-.")
ax.plot(x,x**2, color="orange", lw=.75, linestyle=":")
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x,x+2, color="black", marker="+", markersize=11)
ax.plot(x,x*2, color="blue", marker="o", markersize=15)
ax.plot(x,x**2, color="orange", marker="s", markersize=19)
```

```python
plt.figure(figsize=(12,5))
plt.suptitle("Linestyle Demo", fontsize=16, y=1.1)


plt.subplot(2,2,1)
plt.plot(x,x+2, color="black", marker="+", markersize=11)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("First")

plt.subplot(2,2,2)
plt.plot(x,x*2, color="blue", marker="o", markersize=15)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Second")


plt.subplot(2,2,3)
plt.plot(x,x**2, color="orange", marker="s", markersize=19)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Third")

plt.tight_layout()
plt.show()
```

```python
color = "#FF00FF" #RGB hex code
plt.figure(figsize=(15,5))
plt.suptitle("LineStyles & Markers & Markersize & Color",fontsize=16, y = 1.2)
plt.subplot(2,2,1)
plt.plot(age, salary, color, ls="-", marker="+" ,markersize=20)
plt.subplot(2,2,2)
plt.plot(age, salary_2, "b",ls="-.", marker="o" ,markersize=20)
plt.subplot(2,2,3)
plt.plot(age, salary, "g",ls=":", marker="x" ,markersize=20)
plt.subplot(2,2,4)
plt.plot(age, salary_2, "yellow",ls="--" ,marker="*" ,markersize=20)
plt.tight_layout();
plt.show()
```
